```python
# Initialization
import os
import pylab as py
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
from datetime import date
import warnings
warnings.filterwarnings("ignore")

# Setting color map to the future plots
sns.set_palette('twilight')
```

```python
# Initialization of the data
filepaths = ["./data/" + f for f in os.listdir("./data/") if f.endswith('.csv')]
df_data = pd.concat(map(pd.read_csv, filepaths)).reset_index(drop=True)

# Drop useless data
df_data = df_data.drop(columns=['No'])

df_data
```

### Cleaning the dataset

```python
# Remove unnecessary duplicates
print(f"Raw data: {df_data.shape}")
df_data = df_data.drop_duplicates()
print(f"After removing the duplicates: {df_data.shape}")
```

```python
# Count the number of missing values
df_data.isna().sum()
```

```python
# Solve incomplete data (NA values)

# df_data = df_data.dropna()
# print(f"After removing NA values: {df_data.shape}")
```

```python
# Modify the data types
df_data = df_data.astype({'wd': 'category','station': 'category', 'year': 'category','month': 'category', 'day': 'category','hour': 'category'})
df_data.dtypes
```

```python
fig, axs = plt.subplots(2, 2)
fig.set_size_inches(14, 10)

df_data['year'].value_counts().plot(subplots=True, autopct='%.2f%%', kind='pie', ylabel = "", ax=axs[0][1])
axs[0][1].set_title('YEAR')
df_data['month'].value_counts().plot(subplots=True, autopct='%.2f%%', kind='pie', ylabel = "", ax=axs[1][0])
axs[1][0].set_title('MONTH')
df_data['day'].value_counts().plot(subplots=True, autopct='%.2f%%', kind='pie', ylabel = "", ax=axs[1][1])
axs[1][1].set_title('DAY')
df_data['hour'].value_counts().plot(subplots=True, autopct='%.2f%%', kind='pie', ylabel = "", ax=axs[0][0])
axs[0][0].set_title('HOUR')
```
