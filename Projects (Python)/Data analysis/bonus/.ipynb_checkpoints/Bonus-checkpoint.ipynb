```python
# Initialization
import os
import pylab as py
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
import base64
from datetime import date
from PIL import Image
from IPython.display import HTML
from io import BytesIO
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import validation_curve

sns.set_palette('flare')
RAND_STATE=22
```

```python
# Additional functions for image transforming
pd.set_option('display.max_colwidth', None)
def image_base64(im):
    with BytesIO() as buffer:
        im.save(buffer, 'jpeg')
        return base64.b64encode(buffer.getvalue()).decode()

def image_formatter(im):
    return f'<img src="data:image/jpeg;base64,{image_base64(im)}">'
```

First of all we decided to load all of our data (images and their labels) to the dataframe

```python
data = []
for label in os.listdir('./data'):
    for imgpath in os.listdir('./data/'+label):
        img = Image.open('./data/'+label+'/'+imgpath)
        img.load()
        data.append((img, label))
df = pd.DataFrame(data, columns=['image','label'])
```

Down below are some image examples of each label group

```python
HTML(df.sample(n=5).to_html(formatters={'image': image_formatter}, escape=False))
```

Moreover, for each label group we decided to show average color distribution of all pictures. As we can notice, most of the images have dominant black (0) color, apart from 'CXR' and 'Hand' labels, which contain more gray colors (as their plots are gently offsetted to the right)

As a final step in our EDA analysis, in each label group we counted average value of the same image pixels (among all of the pictures) and displayed newly outcome image underneath. So, basically, each picture generalize the behaviour of the whole label group 

```python
plt.subplots(2,3,figsize=(16,10))
avg_imgs = []
for i,label in enumerate(df['label'].unique()):
    df_hist = df[df['label']==label]['image'].apply(lambda img: np.array(img))
    avg_img = np.zeros((64,64))
    
    for index, img in df_hist.iteritems():
        avg_img = np.add(avg_img, img)           
    avg_imgs.append((avg_img/df_hist.shape[0]).astype(int))
    
    plt.subplot(2,3,i+1)
    plt.xlabel(label)
    plt.imshow(avg_imgs[i], cmap='gray', vmin=0, vmax=255)
```

The examination we made shows us that all of the dataset items (pictures) are of the same nature, have the same size and other properties, so we can start implementing the next (classification) phase.

### PART 2 (CLASSIFICATION) 

```python
# Additional function for image bites transforming
def transform_img(img_data):
    return np.stack(img_data.apply(lambda img: np.array(img).flatten()))
```

First of all we divided our dataset into train (75%) and test (25%) parts

```python
train_img, test_img, train_label, test_label = train_test_split(df['image'], df['label'], test_size=0.25, random_state=RAND_STATE)
```

```python
train_img.shape, test_img.shape
```

#### KNN Algorithm

```python
# Implementation and model training
knn = KNeighborsClassifier(n_jobs=-1)
knn.fit(transform_img(train_img),train_label)
predictions = knn.predict(transform_img(test_img))
```

For a better outcome explanation we displayed some exactness metrics down below. As we can see, implemented model works very well and shows general effectiveness 99-100%

```python
print(classification_report(test_label, predictions))
```

In addition, confusion matrix demonstrates all of mistakes that occured and their influence on the general accuracy outcome. We can notice that some errors during classification happened among 'Hand' and 'CXR' labeling and their accuracy is a bit worse. This may be impacted by the color distribution of an image (as black images have more contrast, so they are easier to classify)

```python
fig, ax = plt.subplots()
fig.set_size_inches(16,8)
labels=test_label.unique()
cm = confusion_matrix(test_label, predictions, labels=labels, normalize='true')
ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=labels).plot(ax=ax, values_format='.3%')
```

Underneath we checked our KNN classifier with different combinations of hyperparameters and got the next conclutions:

- model with the better accuracy has less number of nearest neighbors (K), so that classification depends only on the closest (most similar) labels
- weighted neighbors have a positive influence on the classification algorithm, so that remoted neighbors are less significant than the closest ones (unlike in the uniform method)
- in our case classification doesn't depend on the algorithm it uses (ball_tree, kd_tree or brute)

```python
def plot_validation_curve(ax, title, param_range, train_scores, test_scores):
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    ax.plot(param_range, train_mean, label="Training score", color="darkorange")
    ax.plot(param_range, test_mean, label="Cross-validation score", color="dimgrey")

    ax.fill_between(param_range, train_mean - train_std, train_mean + train_std, color="orange", alpha=0.3)
    ax.fill_between(param_range, test_mean - test_std, test_mean + test_std, color="gainsboro", alpha=0.3)
 
    ax.set_title("Validation Curve - "+title)
    ax.set(xlabel = title, ylabel = "Accuracy")
    ax.legend(loc="best")

plt.subplots(1,3,figsize=(16,10))    
params = {'n_neighbors':[2,3,5],'weights':['uniform', 'distance'],'algorithm': ['ball_tree','kd_tree','brute']}
for index, param_title in enumerate(params):
    ax = plt.subplot(1,3,index+1)
    _train_scores, _test_scores = validation_curve(KNeighborsClassifier(), X=transform_img(test_img), y=test_label, param_name=param_title,
        param_range = params[param_title], scoring="accuracy", cv=3, n_jobs=4, verbose=3)
    plot_validation_curve(ax, param_title, params[param_title], _train_scores, _test_scores)
plt.show()
```
